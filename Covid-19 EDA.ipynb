# Covid-19 Daily Cases Analysis (UK)

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import preprocessing
```

```python
df = pd.read_csv('data/covid-19/uk_covid_19_daily_cases.csv')

df['date'] = pd.to_datetime(df['date'])
df = df.set_index('date')

df = df['2020-02-28':]
```

```python
FIGSIZE = (17, 5)
```
